```python
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import numpy as np
physical_devices = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)

imgs = np.load('/home/tarun/LRCN/ConvLSTM-HAR/fames.npy')
results = np.load('/home/tarun/LRCN/ConvLSTM-HAR/results.npy')
# imgs = tf.squeeze(imgs,axis=-1)
text = 'GeeksforGeeks'
  
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 
  
# org 
org = (0, 50) 
  
# fontScale 
fontScale = 1
   
# Red color in BGR 
color = (0, 0, 255) 

# Line thickness of 2 px 
thickness = 10
for i in len(imgs):
    for t in i:
        z = 'win'
        t = t.reshape(128,128,3)
        t = np.array(t)
        image = cv2.putText(t, str(results[i][0]), org, font, fontScale,  
                 color, thickness, cv2.LINE_AA)
        cv2.imshow(z,cv2.resize(image,(512,512)))
        k = cv2.waitKey(27)
        if k==ord('q'):
            break
```
